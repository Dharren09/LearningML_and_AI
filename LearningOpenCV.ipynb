```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
from IPython.display import Image, display


image_path = r"C:\Users\ARAB-TECH COMPUTERS\OneDrive\Desktop\Aviation\Jey.jpg"

display(Image(filename=image_path))
```

```python
cb_image = cv.imread(image_path, 0)

print(cb_image)
```

```python
print(cb_image.shape)
print(cb_image.dtype)
```

```python
plt.imshow(cb_image)
```

```python
# reversing the channels
cb_image_reversed = cb_image[:, ::-1]
plt.imshow(cb_image_reversed, cmap="gray")
```

```python
# Split color channels
jey_bgr = cv.imread(image_path, cv.IMREAD_COLOR)
b, g, r = cv.split(jey_bgr)

# Show the channels
plt.figure(figsize=[20,5])
plt.subplot(141);plt.imshow(b, cmap='gray');plt.title('Blue Channel');
plt.subplot(142);plt.imshow(g, cmap='gray');plt.title('Green Channel');
plt.subplot(143);plt.imshow(r, cmap='gray');plt.title('Red Channel');

# Merge the channels
jey_bgr_merged = cv.merge((b, g, r))
plt.subplot(144);plt.imshow(jey_bgr_merged[:, :, ::-1]);plt.title('Merged Channels')
```

```python
# Changing color spaces
# dst = cv.cvtColor(image_src, color_code)
# RGB -> Red, Green, Blue

jey_image_rgb = cv.cvtColor(jey_bgr, cv.COLOR_BGR2RGB)
plt.imshow(jey_image_rgb)
```

```python
# change to color spaces from bgr to hsv
# HSV -> Hue, Saturation, Value

jey_hsv = cv.cvtColor(jey_bgr, cv.COLOR_BGR2HSV)

h, s, v = cv.split(jey_hsv)

# ploting individual channels
plt.figure(figsize=[20,5])
plt.subplot(141);plt.imshow(h, cmap="gray");plt.title('H channel');
plt.subplot(142);plt.imshow(s, cmap="gray");plt.title('S channel');
plt.subplot(143);plt.imshow(v, cmap="gray");plt.title('V channel');
plt.subplot(144);plt.imshow(jey_image_rgb);plt.title('Original')
```

```python
# writing and saving images
# cv.imwrite(filename_to_be_stored_in, image[, params])

cv.imwrite('jey_rgb.png', jey_image_rgb)
Image(filename='jey_rgb.png')
```
